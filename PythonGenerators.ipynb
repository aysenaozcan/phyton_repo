```python
# Exercise 1 — Basic Generator
# Write a generator that:
# - Generates numbers from 1 to n, one at a time.
#
# Usage:
# for x in count_up(5):
#     print(x)
#
# Expected output:
# 1
# 2
# 3
# 4
# 5
```

```python
def my_generator(n):

  for value in range(1,n+1):
   yield value

for v in my_generator(5):
    print(v)
```

```python
# Exercise 2 — Even Numbers Generator
# Write a generator that:
# - Generates all even numbers from 0 to n.
#
# Usage:
# list(evens(10))
#
# Expected output:
# [0, 2, 4, 6, 8, 10]
```

```python
def evens(n):
    for value in range(n + 1):
        if value % 2 == 0:
            yield value



list(evens(10))
```

```python
# Exercise 3 — Filter Generator
# Write a generator that:
# - Takes a list as input
# - Yields only odd numbers.
#
# Usage:
# nums = [1, 2, 3, 4, 5, 6]
# list(odd_numbers(nums))
#
# Expected output:
# [1, 3, 5]
```

```python
def odd_numbers(nums):
    for value in nums:
        if value % 2 == 1:
            yield value

nums = [1, 2, 3, 4, 5, 6]
print(list(odd_numbers(nums)))

```

```python
# Exercise 4 — Infinite Generator
# Write a generator that:
# - Generates Fibonacci numbers infinitely.
#
# Usage:
# gen = fibonacci()
# for _ in range(7):
#     print(next(gen))
#
# Expected output:
# 0
# 1
# 1
# 2
# 3
# 5
# 8
```

```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

gen = fibonacci()
for _ in range(7):
    print(next(gen))
```

```python
# Exercise 5 — Real-World Case (Interview Level)
# Write a generator that:
# - Reads a file line by line
# - Uses memory efficiently.
#
# Usage:
# for line in read_file("data.txt"):
#     print(line)
#
# Question:
# Why is using a generator better than readlines() in this case?
```

```python
def read_file(filename):
    with open(filename, "r", encoding="utf-8") as file:
        for line in file:
            yield line.strip()

for line in read_file("data.txt"):
    print(line)

```
