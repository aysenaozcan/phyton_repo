```python
#QUESTION 1 (Easy – *args)

#Write a function that:
#Accepts any number of arguments
#Returns the average of those numbers
#Hint: Use *args
#Function call:
```

```python
def average(*args):
 if len(args) == 0:
        return 0
 return sum(args) / len(args)

average(10, 20, 30)
```

```python
#QUESTION 2 (Medium – **kwargs)

#Write a function that:
#Accepts information using **kwargs
#Prints the information line by line in the following format:
#name: Aysena
#job: Business Analyst
#city: Istanbul
#Function call:
#print_info(name="Ayşena", job="Business Analyst", city="Istanbul")
```

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")


print_info(Name="Aysena", Job="Business Analyst", City="Istanbul")
```

```python
#QUESTION 3 (Medium – *args + **kwargs)

#Write a function that:
#Uses *args for student grades
#Uses **kwargs for settings
#Rules:
#If show=True, print the grades
#If round=2, display the average with 2 decimal places
#Function call:
#calculate(70, 80, 90, show=True, round=2)
#Expected output:
#Grades: (70, 80, 90)
#Average: 80.00
```

```python
def calculate(*args,**kwargs):

    if kwargs.get("show"):
        print(f"Notlar: {args}")
    else:
        print("Notlar yazdırılamaz.")

    average=round(sum(args) / len(args))

    if kwargs.get("round") == 2:
       print(f"{average:.2f}")
    else:

        print(f"Ortalama: {average}")

calculate(70, 80, 90, show=True, round=2)
```
